```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
print(df.head())
df.describe(include='all')
```

Hypothesis:
The key factors deciding the survival of the Titanic disaster were:
- Sex
- Age
- Class (pclass)
- Alone

```python
sns.countplot(data=df, x='sex', hue='survived', palette="Spectral")
plt.title("Survival Count by Sex", fontsize=20)
plt.xlabel("Sex", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.legend(title="Survived", labels=["No", "Yes"])
```

```python
sns.kdeplot(df[df["survived"] == 1]["age"], fill=True, label="Survived", color='green')
sns.kdeplot(df[df["survived"] == 0]["age"], fill=True, label="Did Not Survive", color='red')
plt.title("Age Distribution of Survivors vs Non-Survivors", fontsize=20)
plt.xlabel("Age", fontsize=15)
plt.ylabel("Density", fontsize=15)
plt.legend(fontsize=15)
```

```python
sns.countplot(data=df, x='pclass', hue='survived', palette="Spectral")
plt.title("Survival Count by Passenger Class", fontsize=20)
plt.xlabel("Passenger Class", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.legend(title="Survived", labels=["No", "Yes"])
```

```python
sns.countplot(data=df, x='alone', hue='survived', palette="Spectral")
plt.title("Survival Count by Traveling Alone", fontsize=20)
plt.xticks([0,1], ["With Family", "Alone"], fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.xlabel("")
plt.legend(title="Survived", labels=["No", "Yes"])
```
